```python
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
x_data = [1.0,2.0,3.0]
y_data = [2.0,4.0,6.0]
```

```python
def forward(x,w):
    return x*w 
```

```python
def loss(x,y,w):
    y_predict = forward(x,w)
    return (y-y_predict)**2
```

```python
w = torch.Tensor([1.0])
w.requires_grad = True
lr = 0.01
```

```python
epoch_list = []
loss_list = []
for epoch in range(100):
    seed = np.random.choice(range(len(x_data)))
    loss_val = loss(x_data[seed],y_data[seed],w)
    loss_val.backward()
    w.data -= lr * w.grad.data
    w.grad.data.zero_()
    print(w.data.item())
    epoch_list.append(epoch)
    loss_list.append(loss_val.data.item())
```

```python
plt.plot(epoch_list,loss_list)
plt.xlabel("epoch")
plt.ylabel("loss")
plt.show()
```
