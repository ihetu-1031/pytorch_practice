```python
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
x_data = [1.0,2.0,3.0]
y_data = [7.0,18.0,35.0]
```

```python
def forward(x,w):
    return w[2]*x**2 + w[1]*x + w[0]
```

```python
def loss(x,y,w):
    y_predict = forward(x,w)
    return (y-y_predict)**2
```

```python
w = torch.Tensor([2.0,2.0,2.0])
w.requires_grad = True
lr = 0.01
```

```python
epoch_list = []
loss_list = []
for epoch in range(20000):
    seed = np.random.choice(range(len(x_data)))
    loss_val = loss(x_data[seed],y_data[seed],w)
    loss_val.backward()
    w.data[2] -= lr*w.grad.data[2]
    w.data[1] -= lr*w.grad.data[1]
    w.data[0] -= lr*w.grad.data[0]
    w.grad.data.zero_()
    epoch_list.append(epoch)
    loss_list.append(loss_val.data.item())
    print(w.data)
```

```python
plt.plot(epoch_list,loss_list)
plt.xlabel("epoch")
plt.ylabel("loss")
plt.show()
```
