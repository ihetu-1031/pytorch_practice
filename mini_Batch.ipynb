```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
x_data = np.arange(1.0,200.0,1.0)
y_data = np.arange(2.0,400.0,2.0)
```

```python
def forward(x,w):
    return x*w
```

```python
def cost(x,y,w):
    cost = 0
    for x_val,y_true in zip(x,y):
        y_predict = forward(x_val,w)
        loss_val = (y_true - y_predict)**2
        cost += loss_val
    return cost/len(x)
```

```python
def gradient(x,y,w):
    gradient = 0
    for x_val,y_true in zip(x,y):
        temp = 2*x_val*(x_val*w-y_true)
        gradient += temp
    return gradient/len(x)
```

```python
def get_two(nums):
    index1 = random.randrange(len(nums))
    index2 = random.randrange(len(nums))
    index2 = random.randrange(len(nums)) if index1 == index2 else index2
    return index1,index2
```

```python
w = 1.0
alpha = 0.0001
batch_size = 2
epoch_list = []
loss_list = []
seed = []
x_list = []
y_list = []
print("before: ",forward(400,w))
for epoch in range(1000):
    idx1, idx2 = get_two(x_data)
    seed.append(idx1)
    seed.append(idx2)
    for i in range(batch_size):
        x_list.append(x_data[seed[i]])
        y_list.append(y_data[seed[i]])
    loss_val = cost(x_list,y_list,w)
    gradient_val = gradient(x_list,y_list,w)
    w -= alpha * gradient_val
    print("epoch: ",epoch,"loss: ",loss_val, "w: ", w)
    epoch_list.append(epoch)
    loss_list.append(loss_val)
print("after: ",forward(400,w))    
```

```python
plt.plot(epoch_list,loss_list)
plt.ylabel("loss")
plt.xlabel("epoch")
plt.show()
```
