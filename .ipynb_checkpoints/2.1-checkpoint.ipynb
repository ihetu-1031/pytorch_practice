```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = [1.0,2.0,3.0]
y_data = [2.0,4.0,6.0]
```

```python
def forward(x,w):
    return x*w
```

```python
def loss(x,y,w):
    y_predict = forward(x,w)
    return (y-y_predict)**2
```

```python
w_list = []
mse_list = []
```

```python
for w in np.arange(0,4.1,0.1):
    print("w = ",w)
    l_sum = 0
    for x_val, y_val in zip(x_data, y_data):
        y_predict_val = forward(x_val,w)
        loss_val = loss(x_val,y_val,w)
        l_sum += loss_val
        print('\t', x_val, y_val, y_predict_val, loss_val)
    print("MSE = ", l_sum / len(x_data))
    w_list.append(w)
    mse_list.append(l_sum / len(x_data))
```

```python
plt.plot(w_list,mse_list)
plt.ylabel("MSE")
plt.xlabel("w")
plt.show()
```
